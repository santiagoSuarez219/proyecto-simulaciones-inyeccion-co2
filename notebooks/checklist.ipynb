```python
import os
import pandas as pd

ruta_base = '../data/processed'

registros = []

for caso in os.listdir(ruta_base):
    ruta_caso = os.path.join(ruta_base, caso)

    if not os.path.isdir(ruta_caso):
        continue

    elementos = os.listdir(ruta_caso)

    # Detectar variables automáticamente
    carpetas = [e for e in elementos if e.endswith('layer_cubes') or e.endswith('_tensors')]

    for carpeta in carpetas:
        # Caso especial
        if carpeta == 'layer_cubes':
            var = 'sfd_vd'
        else:
            var = carpeta.replace('_layer_cubes', '').replace('_tensors', '')

        # 🔥 FALTABA ESTO
        ruta_var = os.path.join(ruta_caso, carpeta)

        # contar tensores
        n_archivos = len([
            f for f in os.listdir(ruta_var)
            if f.endswith('.pt')
        ]) if os.path.exists(ruta_var) else 0

        # report json
        report_json = f"{carpeta}_report.json"
        tiene_report = report_json in elementos

        # timeline (puede haber varios)
        timeline = f"timeline_{var}.json"
        tiene_timeline = timeline in elementos

        registros.append({
            "caso": caso,
            "variable": var,
            "carpeta": carpeta,
            "n_tensores": n_archivos,
            "tiene_carpeta": os.path.exists(ruta_var),
            "report_json": tiene_report,
            "timeline_json": tiene_timeline
        })

df = pd.DataFrame(registros)
pivot = df.pivot_table(
    index="caso",
    columns="variable",
    values="n_tensores",
    fill_value=0
)

pivot
```

```python
import os
import shutil

ruta_base = '../data/processed'
caso_base = '001_LatinHyperCube'

ruta_base_caso = os.path.join(ruta_base, caso_base)

# Elementos a copiar
elementos_copiar = [
    'permeability_layer_cubes',
    'porosity_layer_cubes',
    'permeability_layer_cubes_report.json',
    'porosity_layer_cubes_report.json',
    'timeline_permeability.json',
    'timeline_porosity.json'
]

for caso in os.listdir(ruta_base):
    if caso == caso_base:
        continue  # saltar el original

    ruta_destino = os.path.join(ruta_base, caso)

    if not os.path.isdir(ruta_destino):
        continue

    print(f"\nCopiando a: {caso}")

    for elem in elementos_copiar:
        src = os.path.join(ruta_base_caso, elem)
        dst = os.path.join(ruta_destino, elem)

        if not os.path.exists(src):
            print(f"  ❌ No existe en origen: {elem}")
            continue

        # Si es carpeta
        if os.path.isdir(src):
            if os.path.exists(dst):
                print(f"  ⚠️ Ya existe carpeta: {elem}")
            else:
                shutil.copytree(src, dst)
                print(f"  ✔ Carpeta copiada: {elem}")

        # Si es archivo
        else:
            if os.path.exists(dst):
                print(f"  ⚠️ Ya existe archivo: {elem}")
            else:
                shutil.copy2(src, dst)
                print(f"  ✔ Archivo copiado: {elem}")
```
